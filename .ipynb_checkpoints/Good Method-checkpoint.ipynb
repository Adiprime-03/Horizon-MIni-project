```python
from astropy.io import ascii
from astropy.table import Table, Column, join
from numpy import *
import matplotlib.pyplot as plt
import numpy as np
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
data1 = ascii.read('Data/data1.nzpcat', data_start = 0, header_start = 0, delimiter = ' ')
data2 = ascii.read('Data/data2.cat', data_start = 0, header_start = 0, delimiter = ' ')
```

```python
print(data1)
```

```python
print(data2)
```

```python
l1r = data1['ra']
l1d = data1['dec']
l2r = data2['ra']
l2d = data2['dec']
```

```python
coords1 = SkyCoord(l1r, l1d, frame = 'icrs', unit = 'deg')
coords2 = SkyCoord(l2r, l2d, frame = 'icrs', unit = 'deg')
Id, dist, dummy = coords1.match_to_catalog_sky(coords2)
r = 0.4/3600
s = (dist > r*u.deg)
Id[s] = -99
cond_1 = Id>0
```

```python
cond_2 = (data1[cond_1]['f_f160w'] > 0)
cond_3 = (data2[Id[cond_1]][cond_2]['f_F160W'] > 0)
data2[Id[cond_1]][cond_2]['f_F160W']
data1[cond_1][cond_2]['f_f160w']
```

```python
y2 = -2.5*log10(data2[Id[cond_1]][cond_2][cond_3]['f_F160W']) + 25
y1 = -2.5*log10(data1[cond_1][cond_2][cond_3]['f_f160w']) + 25
plt.plot(y1, "r-")
plt.plot(y2, "b:")
```

```python
e = y2 - y1
plt.scatter(y1, e, c = "black", alpha = 0.1)
```
